<strong>Equation 1: Lennard-Jones Equation [1][2]:</strong>
### \begin{equation}
    V_{LJ}(r) = 4\epsilon\left[(\frac{\sigma}{r})^{12} - (\frac{\sigma}{r})^6\right]
\end{equation}

<br>

<strong>Equation 2: Extended Exponential Equation from Yang et al [4]:</strong>
### \begin{equation}
    V_{Exp}(r) = \epsilon\left[e^{\alpha(1 - \frac{r}{\sigma})} - \left((\frac{r}{\sigma})^4 - 2(\frac{r}{\sigma})^2 + 3\right)\right] e^{(\frac{\alpha}{2})(1 - \frac{r}{\sigma})}
\end{equation}

First and foremost, required libraries for the assignment should be imported, so that all later code blocks can run well. For this assignment we need 1. Pandas library and 2. Python's built-in math library:

```python
import pandas as pd
import math
```

<hr>
<strong style = "font-size:20px">Task 1</strong>
<hr>

In this task we need to 1. load the target data from a designated .csv file called <strong><i>"CybulskiT1999_Ar2.csv"</i></strong> and 2. clean & process the datas according to the requirements specified in the assigment sheet.

1. To load the data from the file, we have to utilize Pandas library. The imported dataframe will be stored in variable <code>df</code>:

```python
df_read = pd.read_csv("CybulskiT1999_Ar2.csv", header=0, sep=";")

print("\nLoad data:\n\n")
df_read
```

By printing the data set out, we can see that it comprises of 2 columns or types of data, including: R (Å) - distance in Angstrom unit, and V(r) (Hartree) - potential energy of the correspoding interactions in Hartree unit. 

2. Then the data needs to be cleaned, in which any row with either duplicated data or missing data will be removed from the data set. This can also be done with the help of the Pandas library. Note that I also cloned the dataframe with pandas' function <code>copy()</code>, so that I can use the (cleaned) original dataframe later if needed:

```python
# Drop rows with missing values
df_read.dropna(axis=0, how='any', inplace=True)

# Drop duplicate rows
df_read.drop_duplicates(inplace=True)

# Clone dataframe so that the original dataframe can be used later
df = df_read.copy()

# Print the modified DataFrame
print("\nCleaned data:\n\n")
df
```

3. We then have to convert values of the potential energy column from Hartree to kJ/mol. This can be done by multiplying the values with 2625.5 <strong>[3][5]</strong>

```python
# Conversion factor from Hartree to kJ/mol
hartree_to_kJ_per_mol = 2625.5

# Multiplying the potential energy column with the conversion factor
df['V(r) (Hartree)'] *= hartree_to_kJ_per_mol
df.rename({'V(r) (Hartree)' : 'V(r) (kJ/mol)'}, axis='columns', inplace=True)

# Print the modified DataFrame
print('\nConverted data:\n\n')
df
```

<hr>
<strong style = "font-size:20px">Task 2</strong>
<hr>

- In this task we need to compute

  1. the Lennard-Jones potential energy (Equation 1), and
  2. the exponential potential energy (Equation 2) 

  using the given distances.

<br>

- For this assigment Argon atoms are taken into consideration, as such, epsilon, sigma and alpha used in the computation process are those belong to the Argon atoms. They're already available in the assigment paper: Epsilon = 1.178 (kJ/mol), sigma = 3.75 (Å) and alpha = 13.18

```python
# Constances of the Argon atoms
epsilon = 1.178 # kJ/mol
sigma = 3.75 #Å
alpha = 13.18
```

1. Lennard - Jones potential energy
    - For calculation of potential energy using Lennard-Jones equation <strong>[1][2]</strong>, the following function will be utilized, and to convert the result to unit kJ / mol:

```python
# Function for calculating Lennard-Jones potential energy
def lennardJones(distance:float, epsilon:float, sigma:float) -> float:
    return 4 * epsilon * ((sigma / distance) ** 12 - (sigma / distance) ** 6)
```

* The function will then be called on the distance column of the dataframe, adding another column named 'Lennard-Jones potential energy (kJ/mol)'. This can be done by pandas' built-in function <code>apply()</code>:

```python
df['Lennard-Jones potential energy (kJ/mol)'] = df.apply(lambda row: lennardJones(distance=row['R (Å)'], epsilon=epsilon, sigma=sigma), axis=1)
df
```

2. Exponential potential energy
    - For calculation of potential energy using extended exponential Equation from Yang et al <strong>[4]</strong>, the following function will be utilized:

```python
# Function for calculating Lennard-Jones potential energy
def exponentialYang(distance:float, epsilon:float, sigma:float, alpha:float) -> float:
    intermediate_result1 = math.exp((alpha * (1 - distance/sigma)))
    intermediate_result2 = ((distance / sigma) ** 4) - 2 * ((distance / sigma) ** 2) + 3
    intermediate_result3 = math.exp((alpha/2) * (1 - distance/sigma))

    return epsilon * (intermediate_result1 - intermediate_result2 * intermediate_result3)
```

* The function will then again be called on the distance column of the dataframe, adding another column named 'Exponential potential energy (kJ/mol)'. This can be done by pandas' built-in function <code>apply()</code>:

```python
df['Exponential potential energy (kJ/mol)'] = df.apply(lambda row: exponentialYang(distance=row['R (Å)'], epsilon=epsilon, sigma=sigma, alpha=alpha), axis=1)
df
```

- For this task we have to plot these data using lines:
    + the QM target energies
    + the Lennard - Jones energies (Equation 1)
    + the exponential energies (Equation 2)
  with the x-axis being the distances
- I find the library <code>pyplot</code> to be extemely useful for this task, as the graph can be better customized, and the 3 data lines can be shown on the same chart simultaneously. As such I have imported the library:

```python
import matplotlib.pyplot as plt
```

- The required parameter "ax" can then be retrieved using plt's built-in function subplots()
- The data are then plotted as lines, with the x-axis being the distances (R)
- The function <code>legend()</code> was utilized to create a legend for the chart, which tremendously helped with visibility and easiness to understand

From the graph above we can clearly see, that while the Lennard - Jones equation's results varies not far from the actual values when the distances between the atoms are above a certain threshold, they are unprecise when it comes to smaller distances. The new improvised exponential equation by Yang et al managed to considerably reduce the discrepancies between the calculated results and the actual values, resulting in a more accurate representation of data.
